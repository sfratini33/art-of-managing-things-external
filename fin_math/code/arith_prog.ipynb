```python
# Arithmetic progressions
import tkinter as tk
import tkinter.ttk as ttk
import math

root = tk.Tk()
root.title('Arithmetic Progression Calculator')

# Initialize all potentially dynamic global variables to None or an empty state
# This ensures they exist as names across script runs and can be safely checked
# and cleared without errors.
a = []
t = None
t1 = None
s = None
s1 = None
r = None
r1 = None

def quit_app():
    root.destroy()
    
def clear():
    """Robustly clears all input and output fields."""
    global a, t, t1, s, s1, r, r1
    
    # 1. Clear input entry fields and reset their variables to 0
    q1.delete(0, tk.END)
    x.set(0.0)
    
    q2.delete(0, tk.END)
    y.set(0.0)
    
    q3.delete(0, tk.END)
    z.set(0)
    
    # 2. Clear output fields by checking if their widget variables exist and destroying them
    # and then resetting the names to None for the next calculation.
    if t is not None:
        t.destroy()
        t = None
    if t1 is not None:
        t1.destroy()
        t1 = None
    
    if s is not None:
        s.destroy()
        s = None
    if s1 is not None:
        s1.destroy()
        s1 = None
    
    if r is not None:
        r.destroy()
        r = None
    if r1 is not None:
        r1.destroy()
        r1 = None
        
    # 3. Clear the calculated progression list
    a.clear()

def aprog():
    # Function that computes the first n terms of an arithmetic progression and the sum of those terms
    # Removed functools.partial dependency, use global Tkinter variables
    global a, t, t1, s, s1
    start = x.get()
    step = y.get()
    n = z.get()
    
    # Reset the progression list
    a = []
    
    # Compute progression
    next_val = start
    for i in range(n):
        a.append(next_val)
        next_val = next_val + step

    # Compute sum of the progression
    if n > 0:
        # Avoid built-in function override
        prog_sum = round((n/2)*(a[0]+a[n-1]), 3)
    else:
        prog_sum = 0

    # Output the progression
    w = math.trunc(abs(start) + n * abs(step))
    w = len(str(w)) + 6
    if start + n * step > 0: 
        w = 10 * w
    else: 
        w = 10 * (w + 1)

    # 1. Clean up old widgets before creating new ones on top
    if t is not None:
        t.destroy()
    if t1 is not None:
        t1.destroy()
        
    # 2. Create and grid the progression text area and label
    t1 = tk.Label(root, text="The progression is :")
    t1.grid(row=7, column=1, padx=3, pady=11)
    t = tk.Text(root, height=7, width=w)
    t.grid(row=7, column=2, padx=3, sticky='w')

    # Create a scrollbar and associate it with the text widget t
    scrollb = ttk.Scrollbar(root, command=t.yview)
    scrollb.grid(row=7, column=3, sticky='nesw')
    t['yscrollcommand'] = scrollb.set

    # 3. Insert text into the text area
    i = 0
    for item in a:
        i = i + 1
        # Round term for output
        rounded_val = str(round(item, 3))
        if i % 10 != 0 and i != n:
            t.insert(tk.END, rounded_val + ', ')
        elif i != n:
            t.insert(tk.END, rounded_val + '\n')
        else:
            t.insert(tk.END, rounded_val)

    # Output sum of the progression
    if s is not None:
        s.destroy()
    if s1 is not None:
        s1.destroy()
        
    s1 = tk.Label(root, text="The sum of the progression is :")
    s1.grid(row=8, column=1, padx=3, pady=11)
    s = tk.Text(root, height=1, width=w)
    s.insert(tk.END, str(prog_sum))
    s.grid(row=8, column=2, padx=3, sticky='w')

def nth_term():
    # Function that computes the nth term in an arithmetic progression
    # Removed functools.partial dependency, use global Tkinter variables
    global r, r1
    start = x.get()
    step = y.get()
    n = z.get()

    # Compute nth term in progression
    n_term = start + (n-1)*step

    # Print nth term
    # 1. Clean up old widgets before creating new ones on top
    if r is not None:
        r.destroy()
    if r1 is not None:
        r1.destroy()
        
    # 2. Create and grid the new widgets
    r1 = tk.Label(root, text="Term number %s in the progression:" % n)
    r1.grid(row=6, column=1, padx=3, pady=11)
    r = tk.Text(root, height=1, width=11)
    r.insert(tk.END, str(n_term))
    r.grid(row=6, column=2, padx=3, sticky='w')

# Input setup (declaring variables, creating entry widgets)
tk.Label(root, text="Initial term in progression:").grid(row=1, column=1)
x = tk.DoubleVar()
# Two lines for widget creation and layout
q1 = tk.Entry(root, textvariable=x)
q1.grid(row=1, column=2, padx=3, sticky='w')

tk.Label(root, text="Length of a step:").grid(row=2, column=1)
y = tk.DoubleVar()
q2 = tk.Entry(root, textvariable=y)
q2.grid(row=2, column=2, padx=3, sticky='w')

tk.Label(root, text="Number of terms (n):").grid(row=3, column=1)
z = tk.IntVar()
q3 = tk.Entry(root, textvariable=z)
q3.grid(row=3, column=2, padx=3, sticky='w')

# Button setup (use functions as commands directly, no partials needed)
button1 = tk.Button(root, text="Generate Progression and its Sum", command=aprog)
button1.grid(row=4, column=1, padx=3, pady=3)

button2 = tk.Button(root, text="Generate nth term in Progression", command=nth_term)
button2.grid(row=4, column=2, padx=3, pady=3, sticky='w')

button3 = tk.Button(root, text='Clear', command=clear)
button3.grid(row=11, column=1, pady=5)

button4 = tk.Button(root, text='Quit', command=quit_app)
button4.grid(row=11, column=2, padx=3, pady=5, sticky='w')

# Run the application
if __name__ == '__main__':
    root.mainloop()
```
