# 8.5.1 Stock Valuation Calculator

**File name:** stock_valuation.ipynb

This calculator supports stock valuation under constant dividends, the Gordon Growth Model, a required-return decomposition, and the price-to-earnings ratio (assuming 100% payout so that the next dividend equals next-period earnings). On each tab, choose what to solve for from the drop-down list, enter the remaining inputs, and click **Compute**. For growth models, enter only *D*₀ or *E*₀; *D*₁ and *E*₁ are computed as *D*₀(1 + *g*) and *E*₀(1 + *g*).

**Constant Dividend tab** (perpetuity *P* = *D*/*r*):

- Compute **price** *P*: constant dividend *D*, required return *r* (decimal per period).
- Compute **required return** *r*: *D*, *P*.
- Compute **dividend** *D*: *r*, *P*.

**Gordon Growth tab** (*P* = *D*₁/(*r* − *g*), *D*₁ = *D*₀(1 + *g*)):

- Compute **price** *P*: *D*₀, growth rate *g*, *r*.
- Compute **required return** *r*: *D*₀, *g*, *P*.
- Compute **growth rate** *g*: *D*₀, *r*, *P*.
- Compute **prior dividend** *D*₀: *g*, *r*, *P*.

**Return Decomposition tab** (*r* = *D*₁/*P* + *g*):

- Report dividend yield *D*₁/*P*, capital-gains yield *g*, and total return *r*: *D*₀, *g*, *P*.

**P/E Ratio tab** (full payout: *D*₁ = *E*₁, *P*/*E*₁ = 1/(*r* − *g*)):

- **Show all** (*P*/*E*, earnings yield, model price): *E*₀, *g*, *r*, *P*.
- Compute **price** *P*: *E*₀, *g*, *r*.
- Compute **required return** *r*: *E*₀, *g*, *P*.
- Compute **growth rate** *g*: *E*₀, *r*, *P*.
- Compute **prior earnings** *E*₀: *g*, *r*, *P*.

**How to use:**

- Open the **P/E Ratio** tab (or any other tab), select the desired calculation from **Solve for**, fill in the other fields, and click **Compute**. Results appear in the text area below the inputs.
- Click **Clear** to reset the inputs and the result area on that tab.
- Rates *r* and *g* must be entered as **decimals per period** (e.g. 0.08 for 8% per period), not as percentages (do not type 8 or 8%).
